```python
from ultralytics import YOLO

model = YOLO('./training/yolov8/best.pt')

# results = model.predict('/content/drive/MyDrive/Final_Project/dataset/train_img/video_test/test_video_1min.mp4', save = True)
results = model.predict('/content/drive/MyDrive/1min_videos/7/vid_20201125_2000_reset.mp4', save = True)


for frame_idx, result in enumerate(results):
    object_count = len(result.boxes)  # 計算當前幀的物件數量
    print(f"Frame {frame_idx}: {object_count} objects detected")

    # 額外印出每個物件的資訊
    for box in result.boxes:
        x1, y1, x2, y2 = box.xyxy[0]  # 取得 bounding box 左上角與右下角座標
        conf = box.conf[0]  # 置信度
        cls = box.cls[0]  # 類別索引
        print(f"  Class: {int(cls)}, Confidence: {conf:.2f}, Box: [{x1:.2f}, {y1:.2f}, {x2:.2f}, {y2:.2f}]")
```

✅ 依 frame_數字排序 讀取影像

✅ 使用 YOLOv8 進行物件偵測

✅ 輸出標註後影像 至 output_folder

✅ 輸出偵測結果 TXT 檔案 (image_name, class, confidence, x, y, width, height)

✅ 可調整 confidence > 0.5 來減少重複偵測

你可以更換 yolov8n.pt 變成 yolov8s.pt、yolov8m.pt 來使用更強大的模型。

```python
import os
import cv2
import torch
from ultralytics import YOLO

# 參數設定
model_path = "./training/yolov8/best.ptt"  # 你的 YOLOv8 模型權重檔案
input_folder = "./data/video"  # 輸入影像資料夾
output_folder = "/content/drive/MyDrive/Final_Project/dataset/position_research/yolov8_feed_output"  # 偵測後影像輸出資料夾
output_txt = "/content/drive/MyDrive/Final_Project/dataset/position_research/Yolov8_output_test_image.txt"  # 偵測結果輸出

# # overlap實驗使用
# input_folder = "/content/drive/MyDrive/Final_Project/dataset/overlap_test/overlap_img"
# output_folder = "/content/drive/MyDrive/Final_Project/dataset/overlap_test/overlap_yolov8_output_image"
# output_txt = "/content/drive/MyDrive/Final_Project/dataset/overlap_test/overlap_Yolov8_output_image.txt"



# 確保輸出資料夾存在
os.makedirs(output_folder, exist_ok=True)

# 讀取 YOLOv8 模型
model = YOLO(model_path)

# 取得影像檔案並依照 frame_ 數字排序
image_files = sorted(
    [f for f in os.listdir(input_folder) if f.endswith((".jpg", ".png", ".jpeg"))],
    key=lambda x: int(x.split("_")[-1].split(".")[0])  # 根據 frame_X 數字排序
)

# 建立輸出 TXT 檔案
with open(output_txt, "w") as f_out:
    for img_file in image_files:
        img_path = os.path.join(input_folder, img_file)
        img = cv2.imread(img_path)
        height, width, _ = img.shape

        # 進行物件偵測
        results = model(img)

        # 解析 YOLOv8 偵測結果
        f_out.write(f"Image: {img_file}\n")

        for result in results:
            for box in result.boxes:
                x1, y1, x2, y2 = map(int, box.xyxy[0])  # 取得邊界框座標
                confidence = float(box.conf[0])  # 信心度
                class_id = int(box.cls[0])  # 類別索引
                label = model.names[class_id]  # 類別名稱

                if confidence > 0.75:  # 設定信心閾值
                    f_out.write(f"{label} {confidence:.2f} {x1} {y1} {x2-x1} {y2-y1}\n")

                    # 在影像上標記
                    color = (0, 255, 0)
                    cv2.rectangle(img, (x1, y1), (x2, y2), color, 2)
                    cv2.putText(img, f"{label} {confidence:.2f}", (x1, y1 - 10),
                                cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

        f_out.write("\n")

        # 儲存偵測後影像
        output_img_path = os.path.join(output_folder, img_file)
        cv2.imwrite(output_img_path, img)

print("物件偵測完成，結果已儲存於:", output_txt)
```

計算fps

```python
import time
from ultralytics import YOLO

model = YOLO("yolov8n.pt")  # 使用 YOLOv8 模型
img_path = "test.jpg"  # 替換為你的測試圖片

# 測試 100 次推理時間
num_tests = 100
total_time = 0

for _ in range(num_tests):
    start_time = time.time()
    model(img_path)  # 執行推理
    total_time += time.time() - start_time

# 計算平均 FPS
avg_time_per_image = total_time / num_tests
fps = 1 / avg_time_per_image
print(f"Average FPS: {fps:.2f}")
```
